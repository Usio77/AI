```python
# Importing essentials
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading file path into dataframe object
excel_file_path = 'D:\\GitHubProject\\AI\\Lab_2\\cancer_patient.xlsx' # Dataset's path (Use properties to distinguish betn csv and xlsx file)
df = pd.read_excel(excel_file_path) #pip install openpyxl (It is essential for excell type of files)
```

```python
df.head(5) # To view the first five rows of dataframe
```

```python
sns.countplot(data=df, x='Level') # Shows the distribution of each unique value in level
```

```python
df['Gender'].value_counts() # Counts the no of different values in Gender.
```

```python
# Creating pie chart of values in level
df['Level'].value_counts().plot(kind='pie',autopct='%0.2f',colors = ['red','green','yellow','blue']) # autopct ensures the percentage is displayed in the chart
```

```python
# Histogram
plt.hist(df['Age']) # histogram of age
```

```python
# Plot the distribution of univarient set of observation (age)
sns.distplot(df['Age']) # distribution plot
```

```python
# Box plot of age
sns.boxplot(df['Age'])
```

```python
df.describe() #  provides descriptive statistics of the numerical columns in used DataFrame
```

```python
# Bivarient analysis
```

```python
# Scatter plot
sns.scatterplot(x='Balanced Diet', y='Obesity',hue='Gender' , style='Fatigue',size='Age' ,data=df) 
# x = x axis ; y = y axis ; 
# hue = scatter plot will be colored according to it
# style = points will be styled according to its category
# size = size will be determined with the values of this category
plt.show()
```

```python
# Bar plot
sns.barplot(x='Fatigue',y='Age',hue='Gender',data=df)
```

```python
# Box plot
sns.boxplot(x='Gender',y='Age',hue='Fatigue',data=df)
```

```python
# Displot
sns.distplot(df[df['Gender']==1]['Age'],hist=False)
sns.distplot(df[df['Gender']==2]['Age'],hist=False)
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df['Weight Loss'],df['Gender']))
```
